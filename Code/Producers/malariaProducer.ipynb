```
import requests
import json
from pyspark.sql.types import *
response = requests.get("https://ghoapi.azureedge.net/api/SDGMALARIA?$filter=SpatialDimType%20eq%20%27Country%27")
malariaIncidence = response.json()['value']
dfMalaria = spark.createDataFrame(malariaIncidence, schema = StructType([StructField("SpatialDim", StringType(), True),
                                                                         StructField("TimeDim", StringType(), True),
                                                                         StructField("NumericValue", StringType(), True)]))

# The API only returns things as strings, thus we need to cast these to floats
dfMalaria = dfMalaria.withColumn("TimeDim", dfMalaria["TimeDim"].cast('int')) \
                     .withColumn("NumericValue", dfMalaria["NumericValue"].cast('int')) \
                     .withColumnRenamed("NumericValue","MalariaRatePer1000") \
                     .withColumnRenamed("SpatialDim","Country") \
                     .withColumnRenamed("TimeDim","Year")

display(dfMalaria)
```
